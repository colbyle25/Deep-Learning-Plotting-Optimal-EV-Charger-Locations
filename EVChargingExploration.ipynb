# EV Charging Station Exploration

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
evdata = pd.read_csv('alt_fuel_stations (Mar 18 2025).csv')
```

```python
geoevpoints = gpd.GeoDataFrame(evdata, geometry = gpd.points_from_xy(evdata.Longitude, evdata.Latitude), crs = "EPSG:4326") 
```

```python
va_map = gpd.read_file('virginiamap.shp')
```

```python
fig, ax = plt.subplots(figsize=(40, 40), dpi=600)  # Adjust DPI for resolution
#va_map.plot(ax=ax, color='lightgray', edgecolor='black')  # Virginia map
geoevpoints.plot(ax=ax, markersize=5, color='red', alpha=1, label="EV Stations")  # Plot stations\
plt.title("EV Charging Stations in Virginia")
plt.legend()
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(False)
```
