# Gas Station Density

```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Point
```

```python
gas_stations = gpd.read_file('../Data/gasstations.geojson')
```

```python
fig, ax = plt.subplots(figsize=(40, 40), dpi=600)
gas_stations.plot(ax=ax, markersize=5, color='red', alpha=1, label="Gas Stations")
plt.title("Gas Stations in Virginia")
plt.legend()
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(False)
```
